```python
import pandas as pd
import time
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
import numpy as np
import pymysql as pms
db = pms.connect("140.118.126.136", "123", "1234567890", "test",cursorclass=pms.cursors.DictCursor)
cursor = db.cursor()

cursor.execute("select * from btc_usd order by DATE DESC")
result = cursor.fetchall()
btc_data = pd.DataFrame(result)
btc_data = btc_data.assign(DATE=pd.to_datetime(btc_data['DATE']))
btc_data['VOLUME'] = (pd.to_numeric(btc_data['VOLUME'], errors='coerce').fillna(0))
btc_data['VOLUME'] = btc_data['VOLUME'].astype('int64')
btc_data
```

```python
#def some fun tp process data
def add_newcol(btc_data): #新增closeoffhigh和volatolity欄位，增加模型訓練的準確度
    market_info = btc_data[btc_data['DATE']>='2017-01-01'] #將資料只獲取從2017-01-01之後的內容，存到 market_info
    kwargs = { 'CLOSE_OFF_HIGH': lambda x: 2*(x['HIGH'] - x['LAST']) / (x['HIGH'] - x['LOW']) - 1,
          'VOLATILITY': lambda x: (x['HIGH'] - x['LOW']) / (x['MID'])} 
    market_info = market_info.assign(**kwargs)
    return market_info
def create_model_data(btc_data):#選取待會訓練模型所需要的資料
    model_data = btc_data[["DATE"]+["LAST"]+["VOLUME"]+["CLOSE_OFF_HIGH"]+["VOLATILITY"]]
    model_data = model_data.sort_values(by='DATE')
    return model_data
def create_input_data(data,window_len):#決定訓練時以多少天的大小作訓練(window_len)，以每個為window_len大小的array，作為待會訓練模型的input
    norm_cols = ['LAST','VOLUME']
    inputs = [] #將Close,Volume以每筆的第一個資料來作正規化，讓值介於-1,1之間
    for i in range(len(data)-window_len):
        temp_set = data[i:(i+window_len)].copy()
        for col in norm_cols:
            temp_set.loc[:, col] = temp_set[col]/temp_set[col].iloc[0] - 1 
        inputs.append(temp_set)
    return inputs
def create_output_data(data,window_len):#模型的輸出
    return (data['LAST'][window_len:].values / data['LAST'][:-window_len].values) - 1
def data_to_np(data):#原本資料的型態為dataframe，這裡將型態轉為numpy array，模型的資料是以這種型態為主
    data = [np.array(datas)for datas in data]
    data = np.array(data)
    return data
```

```python
split_date = '2017-06-01' #將training_set,test_set從這個日期做劃分
training_set, test_set = model_data[model_data['DATE']<split_date], model_data[model_data['DATE']>=split_date]
test_set.head()
```

```python
training_set = training_set.drop('DATE', 1) #將Date的欄位刪掉，因為之後不會需要用到它，因為後面要將形式轉為np array，故只留數值的部分
test_set = test_set.drop('DATE', 1)
```

```python
window_len = 10 #決定模型要獲取幾天前的數據，隨意選擇
norm_cols = ['LAST','VOLUME']
LSTM_training_inputs = create_input_data(training_set,window_len)
LSTM_test_inputs = create_input_data(test_set,window_len)
LSTM_training_outputs = create_output_data(training_set,window_len)
```

```python
import tensorflow as tf
class LSTMRNN(object):
    def __init__(self,input_data,output_data,input_size,output_size,batch_size,n_steps,cell_size):
        self.input_size=input_size
        self.output_size=output_size
        self.batch_size=batch_size
        self.step_size=n_steps
        self.cell=cell_size
        self.inputdata = input_data
        self.outputdata = output_data
        with tf.variable_scope('in_hidden'):
            self.add_input_layer()
        with tf.variable_scope('LSTM_cell'):    
            self.add_cell()
        with tf.variable_scope('out_hidden'):
            self.add_output_layer()
        with tf.name_scope('cost'):
            self.compute_cost()
        with tf.name_scope('train'):
            self.opti_pred=tf.train.AdamOptimizer(0.02).minimize(self.loss)
            
    def add_input_layer(self):
        inputdata = tf.reshape(self.inputdata,[-1,self.input_size])
        input_cell = tf.layers.dense(inputdata,self.cell)
        input_cell = tf.reshape(input_cell,[-1,self.step_size,self.cell])
        self.input_cell = tf.layers.dropout(input_cell,0.25)
    def add_cell(self):
        lstm_cell=tf.contrib.rnn.BasicLSTMCell(self.cell,forget_bias=1,state_is_tuple=True)
        with tf.name_scope('initial_state'):
            self.initial_state=lstm_cell.zero_state(self.batch_size,dtype=tf.float32)
        self.output_cell,self.final_state=tf.nn.dynamic_rnn(lstm_cell,self.input_cell,initial_state=self.initial_state,time_major=False)
    def add_output_layer(self):        
        output = tf.reshape(self.output_cell,[-1,self.cell])
        self.pred_out = tf.layers.dense(output,self.output_size)  
    def compute_cost(self):
        self.loss=tf.reduce_mean(tf.square(tf.reshape(self.pred_out,[-1])-tf.reshape(self.outputdata, [-1])))
```

```python
batch_size = 1
cell = 20
input_size = 4
output_size = 1
step_size = 10
epochs = 50
output_pred = []
temp = 1
tf_x = tf.placeholder(tf.float32,[None,step_size,input_size])
tf_y = tf.placeholder(tf.float32,[None,1,output_size])
model=LSTMRNN(tf_x,tf_y,input_size,output_size,batch_size,step_size,cell)
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    saver=tf.train.Saver(tf.global_variables(),max_to_keep=1)
    for i in range(epochs):
        for j in range(len(LSTM_training_inputs)):
            if(i==0):
                _,loss,state,pred = sess.run([model.opti_pred, model.loss, model.final_state, model.pred_out],feed_dict ={tf_x:LSTM_training_inputs[np.newaxis,j],tf_y:LSTM_training_outputs[np.newaxis,j]})
            else:
                _,loss,state,pred = sess.run([model.opti_pred, model.loss, model.final_state, model.pred_out],feed_dict ={tf_x:LSTM_training_inputs[np.newaxis,j],tf_y:LSTM_training_outputs[np.newaxis,j],model.initial_state:state})
        if(loss<temp):
            saver.save(sess,'ckpt/bitcoin.ckpt',global_step=i+1)
            temp = loss
        print(i,": ", round(loss, 4))
    module_file = tf.train.latest_checkpoint('ckpt/')
    saver.restore(sess, module_file)
    for i in range(len(LSTM_test_inputs)):
        prob = sess.run(model.pred_out,feed_dict={tf_x:LSTM_test_inputs[np.newaxis,i]})
        predict = prob.reshape((-1))
        output_pred.append(predict)
output_pred = np.array(output_pred)
```
